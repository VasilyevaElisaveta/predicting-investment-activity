## Индекс мосбиржи полной доходности MCFTR

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("archive_indices_2018-01-01-2023-12-31.csv", encoding="Windows-1251", sep=";")
```

```python
df.head()
```

```python
df = df.drop(columns=['NAME', 'OPEN', 'OPEN', "HIGH", "LOW", "VALUE", "DURATION", "YIELD"])
```

```python
df.head()
```

```python
df['TRADEDATE'] = pd.to_datetime(df['TRADEDATE'])
df.CLOSE = df.CLOSE.str.replace(',', '.').astype(np.float32)
```

```python
df.info()
```

```python
df_2018 = df[df['TRADEDATE'].between('2018-01-01', '2018-12-31')]
df_2019 = df[df['TRADEDATE'].between('2019-01-01', '2019-12-31')]
df_2020 = df[df['TRADEDATE'].between('2020-01-01', '2020-12-31')]
df_2021 = df[df['TRADEDATE'].between('2021-01-01', '2021-12-31')]
df_2022 = df[df['TRADEDATE'].between('2022-01-01', '2022-12-31')]
df_2023 = df[df['TRADEDATE'].between('2023-01-01', '2023-12-31')]
```

```python
df_2019.shape
```

```python
value_2018 = np.mean(df_2018["CLOSE"])
value_2019 = np.mean(df_2019["CLOSE"])
value_2020 = np.mean(df_2020["CLOSE"])
value_2021 = np.mean(df_2021["CLOSE"])
value_2022 = np.mean(df_2022["CLOSE"])
value_2023 = np.mean(df_2023["CLOSE"])
```

```python
value_2018, value_2019, value_2020, value_2021, value_2022, value_2023
```
